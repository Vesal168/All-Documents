```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python

feature_vectors = np.loadtxt(r'E:\Drive D\MA-ICT Convergence\Thesis\Human-Human-Interaction\src\feature_vector\add_one_class_feature_vectors.csv', delimiter=',')
```

```python

labels_df = pd.read_csv(r'E:\Drive D\MA-ICT Convergence\Thesis\Human-Human-Interaction\src\feature_vector\add_one_class_labels.csv')
labels = labels_df['Label'].values
```

```python

X_train, X_test, y_train, y_test = train_test_split(feature_vectors, labels, test_size=0.2, random_state=42)
```

```python

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python

y_pred_rf = rf_classifier.predict(X_test)
```

```python

accuracy_rf = accuracy_score(y_test, y_pred_rf)
classification_report_rf = classification_report(y_test, y_pred_rf)
```

```python
print("Random Forest Accuracy:", accuracy_rf)
print("Random Forest Classification Report:\n", classification_report_rf)
```

```python
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
```

```python

plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="Blues", xticklabels=sorted(set(labels)), yticklabels=sorted(set(labels)))
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix - Random Forest Model')
plt.show()
```
