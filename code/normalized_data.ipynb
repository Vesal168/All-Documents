```python
import os
import numpy as np

# Assuming you have a NumPy array called 'keypoints' with zero values
# Replace 0 values with the worst value (minimal value other than 0) for each column
def replace_zero_values(keypoints):
    min_gt0 = np.ma.array(keypoints, mask=keypoints <= 0).min(0)
    keypoints = np.where(keypoints == 0, min_gt0, keypoints)
    return keypoints

# Path to the main folder containing subfolders
main_folder = r"E:\Drive D\MA-ICT Convergence\Thesis\Human-Human-Interaction\dataset\keypoints_extracted_clean"

# List of subfolder names (class names)
class_names = ['hugging', 'kicking', 'punching', 'pointing', 'none-interaction', 'pushing']

# Loop through each class folder
for class_name in class_names:
    class_folder = os.path.join(main_folder, class_name)
    # Loop through each file in the class folder
    for file_name in os.listdir(class_folder):
        file_path = os.path.join(class_folder, file_name)
        if file_name.endswith('.npy'):
            keypoints = np.load(file_path)
            updated_keypoints = replace_zero_values(keypoints)
            # Save the updated keypoints back to the same file
            np.save(file_path, updated_keypoints)
            print(f"Updated keypoints saved to {file_path}")
```

split data into train and test 70-30

```python
import os
import numpy as np
from sklearn.model_selection import train_test_split

# Define the path to your main folder (keypoint)
main_folder = r"E:\Drive D\MA-ICT Convergence\Thesis\Human-Human-Interaction\dataset\interpolated_keypoints"

# Initialize lists to store data and labels
data = []
labels = []

# Iterate through each subfolder
for subfolder in os.listdir(main_folder):
    subfolder_path = os.path.join(main_folder, subfolder)
    if os.path.isdir(subfolder_path):
        # Load numpy files from each subfolder
        for file in os.listdir(subfolder_path):
            if file.endswith(".npy"):
                file_path = os.path.join(subfolder_path, file)
                keypoints = np.load(file_path)  # Load numpy data
                data.append(keypoints)
                labels.append(subfolder)  # Use subfolder name as label

# Convert lists to numpy arrays
data = np.array(data)
labels = np.array(labels)

# Split data into train and test sets with a 70-30 split
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, random_state=42)

# Now you have X_train, y_train (70% of data), and X_test, y_test (30% of data)
print("Train data shape:", X_train.shape)
print("Y train shape:", y_train.shape)
print("Test data shape:", X_test.shape)
print("Y test shape:", y_test.shape)

output_folder = r"E:\Drive D\MA-ICT Convergence\Thesis\Human-Human-Interaction\dataset\split-data-interpolate"
os.makedirs(output_folder, exist_ok=True)
np.save(os.path.join(output_folder, "X_train.npy"), X_train)
np.save(os.path.join(output_folder, "y_train.npy"), y_train)
np.save(os.path.join(output_folder, "X_test.npy"), X_test)
np.save(os.path.join(output_folder, "y_test.npy"), y_test)

print("Data saved successfully in the 'split_data' folder!")
```

```python
import numpy as np

def calculate_distance(keypoints):
    num_keypoints = keypoints.shape[0]
    distances = np.zeros((num_keypoints, num_keypoints))
    
    for i in range(num_keypoints):
        for j in range(i+1, num_keypoints):
            if np.any(keypoints[i] != 0) and np.any(keypoints[j] != 0):  # Check if keypoints are detected
                distances[i][j] = np.linalg.norm(keypoints[i] - keypoints[j])
            else:
                distances[i][j] = np.nan  # If one of the keypoints is not detected, set distance to NaN
    
    return distances

keypoints = np.load(r"E:\Drive D\MA-ICT Convergence\Thesis\Human-Human-Interaction\dataset\keypoints_extracted_clean\pointing\pointing_5.npy")

distances = calculate_distance(keypoints)
print(distances)
```

pairwise keypoint for feature extract

```python
import numpy as np

# Load the numpy file containing keypoints
keypoints_array = np.load(r"E:\Drive D\MA-ICT Convergence\Thesis\Human-Human-Interaction\dataset\keypoints_extracted_clean\pointing\pointing_5.npy")

# Function to calculate Euclidean distance between two keypoints
def euclidean_distance(keypoint1, keypoint2):
    return np.linalg.norm(keypoint1 - keypoint2)

# Function to calculate Euclidean distances between all pairs of keypoints in a frame
def pairwise_distances(frame_keypoints):
    num_keypoints = frame_keypoints.shape[0]
    distances = np.zeros((num_keypoints, num_keypoints))
    for i in range(num_keypoints):
        for j in range(num_keypoints):
            distances[i, j] = euclidean_distance(frame_keypoints[i], frame_keypoints[j])
    return distances

# Iterate through each frame in the keypoints array
for frame_keypoints in keypoints_array:
    # Calculate pairwise distances between keypoints in the current frame
    distances = pairwise_distances(frame_keypoints)
    print("Pairwise distances for frame:")
    print(distances)
```

```python
import numpy as np

def replace_zeros_with_previous_frame_keypoints(keypoints_array):
    # Iterate through each frame
    for i in range(1, len(keypoints_array)):
        # Iterate through each keypoint in the current frame
        for j in range(len(keypoints_array[i])):
            # If the current keypoint is [0, 0], replace it with the corresponding keypoint from the previous frame
            if np.all(keypoints_array[i][j] == 0):
                keypoints_array[i][j] = keypoints_array[i-1][j]
            # If the current keypoint is detected, update the previous keypoint
            else:
                keypoints_array[i-1][j] = keypoints_array[i][j]
    return keypoints_array

# Test the function
keypoints_array = np.load(r"E:\Drive D\MA-ICT Convergence\Thesis\Human-Human-Interaction\dataset\keypoints_extracted_clean\hugging\hugging_3.npy")
updated_keypoints_array = replace_zeros_with_previous_frame_keypoints(keypoints_array)
print(updated_keypoints_array)

```

```python
import numpy as np

def interpolate_keypoints(keypoints):
    interpolated_keypoints = keypoints.copy()
    num_frames, num_keypoints, _ = keypoints.shape
    
    for frame_idx in range(num_frames):
        for kp_idx in range(num_keypoints):
            if np.isnan(keypoints[frame_idx, kp_idx]).any():
                valid_neighbors = []
                for neighbor_idx in range(max(0, kp_idx - 1), min(num_keypoints, kp_idx + 2)):
                    if not np.isnan(keypoints[frame_idx, neighbor_idx]).any():
                        valid_neighbors.append(keypoints[frame_idx, neighbor_idx])
                
                if valid_neighbors:
                    interpolated_keypoints[frame_idx, kp_idx] = np.mean(valid_neighbors, axis=0)
    
    return interpolated_keypoints

# Example usage
keypoints = np.array([[[100, 20], [200, 0]], [[np.nan, np.nan], [300, 50]]])
interpolated_keypoints = interpolate_keypoints(keypoints)
print(interpolated_keypoints)
```
