```python
import os
import numpy as np

def preprocess_keypoints_folder(keypoints_folder, desired_num_keypoints):
    files = os.listdir(keypoints_folder)
    for file in files:
        filepath = os.path.join(keypoints_folder, file)
        data = np.load(filepath)
        num_keypoints = data.shape[0]
        if num_keypoints < desired_num_keypoints:
            # Pad with zeros
            padded_data = np.pad(data, ((0, desired_num_keypoints - num_keypoints), (0, 0)), mode='constant')
        elif num_keypoints > desired_num_keypoints:
            # Truncate
            padded_data = data[:desired_num_keypoints, :]
        else:
            padded_data = data
        np.save(filepath, padded_data)

# Define parameters
keypoints_folder = r"E:\Drive D\MA-ICT Convergence\Thesis\Human-Human-Interaction\dataset\30_frames_npy\pushing"
desired_num_keypoints = 36

# Preprocess keypoints folder
preprocess_keypoints_folder(keypoints_folder, desired_num_keypoints)
```
